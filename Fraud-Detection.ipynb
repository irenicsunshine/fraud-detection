```python
# Cell 5: Visualize Your Outstanding Results  
def create_performance_visualization(results):
    """Create professional visualizations of your model performance"""
    
    # Performance metrics comparison
    fig = make_subplots(
        rows=2, cols=2,
        subplot_titles=('Performance Metrics Comparison', 'Model Accuracy Comparison',
                       'Precision vs Recall Analysis', 'F1-Score Rankings'),
        specs=[[{"type": "bar"}, {"type": "bar"}],
               [{"type": "scatter"}, {"type": "bar"}]]
    )
    
    # Extract metrics for visualization
    models = list(results.keys())
    metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score', 'AUC']
    
    # 1. Performance metrics comparison
    for metric in metrics:
        values = [results[model][metric] for model in models]
        fig.add_trace(
            go.Bar(name=metric, x=models, y=values, 
                  text=[f'{v:.3f}' for v in values], textposition='auto'),
            row=1, col=1
        )
    
    # 2. Accuracy comparison
    accuracy_values = [results[model]['Accuracy'] for model in models]
    fig.add_trace(
        go.Bar(name='Accuracy', x=models, y=accuracy_values,
              text=[f'{v:.4f}' for v in accuracy_values], textposition='auto',
              marker_color=['gold', 'silver', 'orange']),
        row=1, col=2
    )
    
    # 3. Precision vs Recall
    precision_values = [results[model]['Precision'] for model in models]
    recall_values = [results[model]['Recall'] for model in models]
    
    for i, model in enumerate(models):
        fig.add_trace(
            go.Scatter(x=[precision_values[i]], y=[recall_values[i]], 
                      mode='markers+text', name=model,
                      text=[model], textposition='top center',
                      marker=dict(size=15, color=['blue', 'red', 'green'][i])),
            row=2, col=1
        )
    
    # 4. F1-Score rankings
    f1_values = [results[model]['F1-Score'] for model in models]
    fig.add_trace(
        go.Bar(name='F1-Score', x=models, y=f1_values,
              text=[f'{v:.4f}' for v in f1_values], textposition='auto',
              marker_color=['crimson', 'gold', 'lightblue']),
        row=2, col=2
    )
    
    fig.update_layout(height=800, showlegend=True, 
                     title_text="Fraud Detection Model Performance Analysis")
    fig.show()
    
    return fig

# Create your performance visualization
performance_chart = create_performance_visualization(model_results)
```
